# ARI matrix

```python
ARI_matrix = []
def fact(num):
  f = 1
  for i in range(1, num + 1):
    f = f * i
  
  return f

def comb(n,r):
  com = fact(n) / (fact(r) * fact(n-r))
  return com

def ARI(X,Y):
  c = 0
  ARI_value = 0
  for i in range(len(X)-1):
    for j in range(i,len(X)-1):
      if((X[i] == X[j+1]) &  (Y[i] == Y[j+1]) ):
        c = c + 1
      if((X[i] != X[j+1]) &  (Y[i] != Y[j+1]) ):
        c = c + 1

  ARI_value   =   c / comb(len(X),2)
  return ARI_value
  
```

# Relabaling

```python
def majority(a):
  mode = max(a, key = a.count)
  a = [mode] * len(a)
  return a
```

```python
def relab(list1, list2):
  list3 = list2.copy()
  p = 0

  for i in range(len(list2)):
    if (list2[i] != list2[p]):
      list3[p : i] = majority(list1[p : i])
      p = i
    if (i == len(list2)-1):
      list3[p : ] = majority(list1[p :])  
      p = i


  #print("Reference Dataset: " + str(list1))
  #print("Given     Dataset: " + str(list2))
  list2 = list3
  ARI_matrix.append(list2)
  print("Modified  Dataset: " + str(list2))
```

```python
for k in range(len(array_2d)):
  relab(array_2d[index], array_2d[k])
```

```python
def sort(list1, lablist):  
    # Outer loop for traverse the entire list  
    for i in range(0,len(list1)-1):  
        for j in range(len(list1)-1):  
            if(list1[j]>list1[j+1]):  
                temp = list1[j]  
                list1[j] = list1[j+1]  
                list1[j+1] = temp
                
                temp2 = lablist[j]
                lablist[j] = lablist[j+1]
                lablist[j+1] = temp2

    return list1, lablist
```
